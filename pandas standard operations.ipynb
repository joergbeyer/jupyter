```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.DataFrame({ 'A' : [1,2,-3,7,-5,6,7], 'B' : [4,5,0,-2,10,11,12]})
```

```python
df.head(10)
```

create a new colum, as a calculation:

```python
df['sum'] = df['A']+df['B']
```

```python
df.head(10)
```

```python
df['divide'] = df['A']/df['B']
```

```python
df.head(10)
```

replace all inf, -inf values with NaN

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.head(10)
```

drop all rows with NaN elements. Here: only row 2

```python
df.dropna(inplace=True)
```

```python
df.head(10)
```

drop all rows, that fulfill a criteria, e.g. Colum "B" < 0

```python
df.drop(df[df.B < 0].index, inplace=True)
df.head(10)
```

```python
# remove all rows, where A is odd
df.drop(df[df.A % 2 != 0].index, inplace=True)
df.head(10)
```

```python
df.loc['sum'] = df.sum()
df.head(10)
```

```python
# slicing: all rows without the last one
df[:-1]
```

```python
df.loc['avg'] = df[:-1].mean()
df.head(10)
```
